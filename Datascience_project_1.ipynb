```python
# =======================================
# Thiranex Internship  Task 1
# Dataset Exploration : Iris Dataset
# =======================================

# 1. Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 2. Load Dataset
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
# The dataset has no header row, so we add column names manually
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris = pd.read_csv(url, header=None, names=columns)

# 3. Basic Inspection
print("First five rows:")
display(iris.head())

print("\nDataset Info:")
iris.info()

print("\nSummary Statistics:")
display(iris.describe())

print("\nCheck for Missing Values:")
display(iris.isnull().sum())

# 4. Simple Visualizations
sns.set(style="whitegrid")

# Pairplot of all numeric columns coloured by species
sns.pairplot(iris, hue="species")
plt.show()

# Boxplot example
plt.figure(figsize=(8,5))
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.title("Sepal Length by Species")
plt.show()

# Histogram example
iris['sepal_length'].hist(bins=20, edgecolor='black')
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()

# 5. Correlation Heatmap
plt.figure(figsize=(6,4))
sns.heatmap(iris.drop(columns='species').corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()

# 6. Key Observations (Markdown cells in notebook)
#     Add a markdown cell summarizing patterns you see, e.g.:
#     Petal length/width clearly separate the three species.
#     Sepal measurements overlap more between species.
```

### Observation
* Petal length and width clearly separate the three species.
* Sepal measurements show some overlap.
* No missing values were found in the dataset.
