```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Task 4: Exploratory Data Analysis (EDA) - Heart Disease Dataset (working gist link)

url = "https://gist.githubusercontent.com/trantuyen082001/1fc2f5c0ad1507f40e721e6d18b34138/raw/heart.csv"
df = pd.read_csv(url)

print("âœ… Dataset Loaded Successfully")
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nDataset Info:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())

# Visualizations

# 1. Target variable distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="output", data=df, palette="Set2")
plt.title("Heart Disease Distribution (output)")
plt.xlabel("Disease (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

# 2. Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap of Features")
plt.show()

# 3. Age distribution
plt.figure(figsize=(8, 5))
sns.histplot(df["age"], bins=20, kde=True, color="skyblue")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# 4. Boxplot of age vs heart disease outcome
plt.figure(figsize=(8, 5))
sns.boxplot(x="output", y="age", data=df, palette="Set2")
plt.title("Age vs Heart Disease (by Output)")
plt.xlabel("Output (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()

# 5. Pairplot of selected features
# 5. Pairplot of selected features (fixing column names for this dataset)
sns.pairplot(df[["age", "trtbps", "chol", "thalachh", "output"]],
             hue="output", palette="husl")
plt.suptitle("Pairplot of Key Features", y=1.02, fontsize=14)
plt.show()




```
