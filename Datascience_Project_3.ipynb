```python
# Task 3: Data Visualization - Wine Quality Dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (red wine quality data from UCI repository)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url, sep=';')

print("Dataset Head:")
print(df.head())
print("\nDataset Info:")
print(df.info())

# --------------------
# 1. Histogram for numerical features
# --------------------
df.hist(bins=15, figsize=(15, 10), edgecolor='black')
plt.suptitle("Histograms of Wine Quality Features", fontsize=16)
plt.show()

# --------------------
# 2. Boxplots for outlier detection
# --------------------
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, orient="h")
plt.title("Boxplots of Wine Quality Features", fontsize=14)
plt.show()

# --------------------
# 3. Scatter plot between two features
# --------------------
plt.figure(figsize=(8, 6))
sns.scatterplot(x="alcohol", y="quality", data=df, alpha=0.6)
plt.title("Alcohol vs Wine Quality", fontsize=14)
plt.xlabel("Alcohol Content")
plt.ylabel("Wine Quality")
plt.show()

# --------------------
# 4. Heatmap for correlations
# --------------------
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Features", fontsize=16)
plt.show()
```
